```python
import pandas as pd
import numpy as np
```

# State Acres

```python
state_crops = "State_Crop.csv"
crops = pd.read_csv(state_crops)
crops.head()
```

```python
# Make a copy of original
state_crops = crops.copy()
```

```python
state_crop = state_crops[["state", "Crop", "Planted Acres", "Volunteer Acres", "Failed Acres", "Prevented Acres",
                          "Not Planted Acres", "Planted and Failed Acres"]]
state_crop.head()
```

```python
# aggregate functions for combining acres
agg_func = {"Planted Acres": 'sum', "Volunteer Acres": 'sum', "Failed Acres": 'sum', "Prevented Acres": 'sum',
            "Not Planted Acres": 'sum', "Planted and Failed Acres": 'sum'}

sc_new = state_crops.groupby(state_crops["state"]).aggregate(agg_func)

sc_new.head(60)
```

```python
sc_new = sc_new.rename(columns={"Planted Acres": "planted", "Volunteer Acres": "volunteer", "Failed Acres": "failed", 
                                "Prevented Acres": "prevented", "Not Planted Acres": "not_planted", "Planted and Failed Acres": "planted_and_failed"})
sc_new.head()
```

```python
sc_new.to_csv("completed/state_acres.csv")
```

# Merging State farm counts with acres

```python
farms = "state_farm_counts.csv"
farm_count = pd.read_csv(farms)
farm_count
```

```python
fc = farm_count.drop(columns=["State Code", "County Code", "State County Code"])
fc = fc[["State", "Farm Count"]]
fc
```

```python
ffunc = {"Farm Count": 'sum'}
fc_df = fc.groupby(["State"]).agg(ffunc)
fc_df.head(50)
```

```python
fc_df.to_csv("completed/farm_count.csv")
```

# State's top 8 crops

```python
# Number of acres of top 11 crops from each states
top_crop = "total_crops.csv"
top_crops = pd.read_csv(top_crop)
top_crops.head()
```

```python
# Create copy of original
tc = top_crops.copy()
```

```python
tc = tc.drop(columns=["State Code", "County Code", "Crop Code", "County", "State County Code", "Crop Type", "Intended Use", "Irrigation Practice"])

tc = tc.rename(columns={"State": "state", "Crop": "crop", "Planted Acres": "planted", "Volunteer Acres": "volunteer", "Failed Acres": "failed", 
                                "Prevented Acres": "prevented", "Not Planted Acres": "not_planted", "Planted and Failed Acres": "planted_and_failed"})
```

```python
tc.head()
```

```python
eleven = ("BARLEY", "CORN", "OATS", "RICE", "SORGHUM", "SOYBEANS",
            "SUGAR BEETS", "WHEAT")

tcdf = tc.loc[tc["crop"].isin(eleven)]
#tcdf = tcdf['Planted Acres', 'Volunteer Acres', 'Failed Acres', 'Prevented Acres', 'Not Planted Acres', 'Planted and Failed Acres'].str.replace(',', '').astype(np.float64)
tcdf.dtypes
```

```python
agg_func2 = {"planted": 'sum', "volunteer": 'sum', "failed": 'sum', "prevented": 'sum',
            "not_planted": 'sum', "planted_and_failed": 'sum'}
tc_df = tcdf.groupby(["state", "crop"]).aggregate(agg_func2)
tc_df.head(11)
```

```python
tc_df.to_csv("completed/state_total_crops.csv")
```
